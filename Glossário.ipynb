## Glossário


- as: alias, apelido
- inplace: realizar operacao
- True: Verdadeiro
- False: Falso
- ==: Igual a
- !=: Diferente de
- ">" (sem aspas): Maior que
- <: Menor que
- ">=" (sem aspas): Maior ou igual a
- <=: Menor ou igual a
- &: e
- |: ou
- on: baseado em qual valor
- how: como, qual tipo de juncao quer fazer

**Tipos de Dados:**
- str: texto
- int: número inteiro (1)
- float: número decimal (1.0, 1.2)
- bool: Verdadeiro/Falso

**Estruturas de Dados:**
- Series: série
- DataFrame: Tabela
- Index: índice
- Array: matriz

**Estruturas de Controle:**
- for: para
- if: se
- else: senão

**Intervalos:**
- range: intervalo

**Operadores:**
- +: adição
- -: subtração
- *: multiplicação
- /: divisão
- %: módulo (resto da divisão)
- **: exponenciação

**Funções:**
- print(): imprimir valor, mostrar valor
- len(): comprimento
- type(): tipo
- sum(): soma
- mean(): média
- max(): máximo
- min(): mínimo
- unique(): valores únicos
- sort_values(): ordenar valores
- groupby(): agrupar por
- def: definir (para criar uma função)
- return: retornar (para retornar um valor de uma função)

**Pandas:**
- read_csv(), read_excel(): ler um arquivo
- to_csv(), to_excel(): salvar um arquivo
- drop(): remover coluna ou linha
- fillna(): preencher valores em branco
- merge(): mesclar com outra base
- pivot_table(): tabela dinâmica
- plot(): gráfico
- loc[]: acessar informacao baseada na coluna
- iloc[]: acessar informacao baseada na linha

**NumPy:**
- array(): criar matriz
- shape: forma
- reshape(): remodelar
- zeros(): matriz de zeros
- ones(): matriz de uns
- arange(): criar intervalo
- linspace(): criar intervalo linear
- random.rand(): matriz de números aleatórios
- dot(): produto de ponto (produto interno)
- transpose(): transpor
